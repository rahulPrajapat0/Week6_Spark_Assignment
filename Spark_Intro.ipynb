```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, expr

spark = SparkSession.builder \
    .appName("Week6_Data_Pipeline") \
    .getOrCreate()

#Read CSV
df = spark.read.csv(
    "data.csv",
    header=True,
    inferSchema=True
)

print("CSV Read Successfully")

#Rename Column
df = df.withColumnRenamed(
    "Customer Name",
    "Customer_Name"
)

#Cast Data Types
df = df.withColumn(
    "Sales",
    expr("try_cast(Sales as double)")
)

df = df.withColumn(
    "Quantity",
    expr("try_cast(Quantity as int)")
)

df = df.withColumn(
    "Discount",
    expr("try_cast(Discount as double)")
)

#Add New Column
df = df.withColumn(
    "Discounted_Sales",
    col("Sales") * (1 - col("Discount"))
)

#Filter Data
df = df.filter(col("Profit") > 500)

# Select Required Columns

final_df = df.select(
    "Customer_Name",
    "Category",
    "Sub-Category",
    "Sales",
    "Discount",
    "Discounted_Sales",
    "Profit"
)

# Show Output

final_df.show(10)


#Save as CSV

final_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("pipeline_csv")

# Save as Parquet
final_df.write \
    .mode("overwrite") \
    .parquet("pipeline_parquet")

print("Data Pipeline Executed Successfully!")
```
